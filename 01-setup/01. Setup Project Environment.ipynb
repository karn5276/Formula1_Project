## Configure Access to Cloud Storage  via Unity Catalog

**Access Cloud Storage(Storage Container)
**


```python
%fs ls 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/'
```

**Create External Location For Storage Container**

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrickskrogerextdl_formula1
URL 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `kroger-dev-sc`)
COMMENT 'External Location For formula1 Container';
```

## Create Catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS formula1
MANAGED LOCATION 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/'
COMMENT 'This is the main catalog for my project';
```

```python
%sql
SHOW CATALOGS
```

## CREATE SCHEMA landing,bronze,silver,gold

```python
%sql
CREATE SCHEMA IF NOT EXISTS formula1.landing; -- here we are not creating any managed location b'coz we don't have any tables in landing, we have just raw files in volume
CREATE SCHEMA IF NOT EXISTS formula1.bronze
MANAGED LOCATION 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1.silver
MANAGED LOCATION 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1.gold
MANAGED LOCATION 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/gold';
```

```python
%sql
use catalog formula1;
show schemas;
```

## CREATE VOLUME FILES

```python
%sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@databrickskrogerextdl.dfs.core.windows.net/landing'
COMMENT 'landing zone for all files' 
```
